```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binned_statistic
df=pd.read_csv('data/2013-2014.csv',sep=',')
df.rename(columns={'Aggregate Sentence Length': 'length'},inplace = True)
dfi=df[(df['Sentence Type']=='INDETERMINATE')]
df=df[(df['Sentence Type']=='DETERMINATE')]
print(df.columns)
```

```python
dfM=df[(df['Gender']=='MALE')]
dfF=df[(df['Gender']=='FEMALE')]
ax = dfM['Age'].plot.hist(bins=25, rwidth=0.9)
plt.xlabel('Age (mean= '+str(round(dfM['Age'].mean(),2))+')',fontsize='13')
plt.title('Age distribution of Male Offenders in Canada in 2013-2014 Seving 2 or More Years')
plt.show()
ax = dfF['Age'].plot.hist(bins=25, rwidth=0.9)
plt.xlabel('Age (mean= '+str(round(dfF['Age'].mean(),2))+')',fontsize='13')
plt.title('Age distribution of Female Offenders in Canada in 2013-2014 Seving 2 or More Years')
plt.show()
```
