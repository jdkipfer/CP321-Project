```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binned_statistic
df=pd.read_csv('data/2013-2014.csv',sep=',')
df.rename(columns={'Aggregate Sentence Length': 'length'},inplace = True)
dfi=df[(df['Sentence Type']=='INDETERMINATE')]
df=df[(df['Sentence Type']=='DETERMINATE')]
print(df.columns)
```

```python
ax = df['Age'].plot.hist(bins=25, rwidth=0.9)
plt.xlabel('Age (mean= '+str(round(df['Age'].mean(),2))+')',fontsize='13')
plt.title('Age distribution of Offenders in Canada in 2013-2014 Seving 2 or More Years')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

y = dfcustody['length']
outliers = y.between(y.quantile(.001), y.quantile(.99))
#df = pd.DataFrame(dict(x=x, y=y, label=labels))

groups = dfcustody[outliers].groupby('Static/Risk')



fig, ax = plt.subplots()
#print(group)
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for name, group in groups:
    ax.plot(group.Age, group.length, marker='o', linestyle='', ms=3, label=name)
ax.legend()

plt.show()
```
